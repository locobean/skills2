```python
import pandas as pd

#load in new data
df = pd.read_csv("messy_cars.csv")
df.head(20)

# droping rows if the value is missing


df_dropped = df.dropna()

df_dropped.head(20)
```

```python
# drop or fill missing values appropriatly
df['Brand'] = df['Brand'].fillna('Unknown_Brand')

df['Money'] = df['Money'].fillna(df['Money'].mean())

df['Miles'] = df['Miles'].fillna(df['Miles'].mean())

df['Color'] = df['Color'].fillna('Unknown_Colour')

df.head(20)




```

```python
# filter the dataset for a meaningful sunset

subset = df.loc[(df["Brand"] == "BMW") & (df["Color"] == "Red")]

print(subset)


```

```python
# sort the data by cost or money
print(df.sort_values(by='Miles', ascending=False))
```

```python
# 1. Drop or fill missing values

# Fill missing 'Brand' and 'Color' with 'Unknown', 'Price' and 'Mileage' with median

df['Brand'] = df['Brand'].fillna('Unknown')

df['Color'] = df['Color'].fillna('Unknown')

df['Money'] = df['Money'].fillna(df['Money'].median())

df['Miles'] = df['Miles'].fillna(df['Miles'].median())

 

# 2. Filter dataset: e.g., cars newer than 2015 with Price > 30000

subset = df[(df['Yr'] > 2015) & (df['Money'] > 30000)]

 

# 3. Sort the data by Price descending

sorted_df = subset.sort_values(by='Money', ascending=False)

 

# 4. Rename columns for clarity

df_clean = df.rename(columns={

    'Car_num': 'CarID',

    'Brand': 'Brand',

    'Mod': 'Model',

    'Yr': 'Year',

    'Money': 'Price',

    'Miles': 'Mileage',

    'Color': 'Color'

})

 

# Show results

sorted_df.head(), df_clean.head()
```

```python
# GROUPBY
# BROUP BY BRAND
brand_totals = df_clean.groupby("Brand")["Mileage"].mean()
brand_totals.head()
```

```python
brand_totals = df_clean.groupby("Brand")["Mileage"].mean().reset_index(name="Average Mileage")
brand_totals.head()
```

```python
#multiple groupbys - aggregations

grouped_mod_color = df_clean.groupby(["Model", "Color"]).agg(
    avg_prices=("Price", "mean"),
    total_cars=("CarID", "count")

).reset_index()

grouped_mod_color.head(20)


```
