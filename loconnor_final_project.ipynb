```python
import pandas as pd # import Pandas

housing = pd.read_csv('day5/housing.csv') # import data
#df.head()
```

```python
import seaborn as sns # import Seaborn
# Median income v median house value
sns.scatterplot(data=housing, x='median_income', y='median_house_value', hue='ocean_proximity', alpha=0.5)
```

```python
# Ocean proximity v median house value
sns.boxplot(data=housing, x='ocean_proximity', y='median_house_value')
```

```python
# Median income for homeowners 

sns.histplot(data=housing, x='median_income', bins=30, kde=True)

```

```python
sns.barplot(data=housing, x='ocean_proximity', y='median_house_value')
```

```python
#house value & location 
sns.scatterplot(
    data=housing, x='longitude', y='latitude',
    hue='median_house_value', palette='viridis', alpha=0.5
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Use only numeric columns
numeric_cols = housing.select_dtypes(include='number')

# Plot heatmap of correlations
sns.heatmap(numeric_cols.corr(), annot=True, cmap='YlGnBu')
plt.show()
```
